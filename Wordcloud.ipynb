# <font color='Dark blue'>Batman Wordcloud</font>

### _Step 1 - Importing libraries_

```python
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt #to display our wordcloud
from PIL import Image #to load our image
import numpy as np #to get the color of our image
```

### _Step 2 - Opening text script of batman_

```python
#Content-related
text = open('batman.txt', 'r').read()
stopwords = set(STOPWORDS)
```

### _Step 3 - Setting the appearance of wordcloud_

```python
#Appearance-related

#Black and White
bnw_mask = np.array(Image.open('batman.png')) 
wc1 = WordCloud(background_color = 'white',
               stopwords = stopwords,
               mask = bnw_mask,
               max_words = 500,
               contour_width = 5,
               contour_color = 'black')

#Color
color_mask = np.array(Image.open('batman.png')) 
wc2 = WordCloud(background_color = 'orange',
               stopwords = stopwords,
               mask = color_mask,
               max_words = 500,
               contour_width = 5,
               contour_color = 'purple')
```

### _Step 4 - Filling colours and converting to wordcloud_

```python
#For black and white
wc1.generate(text)
image_colors = ImageColorGenerator(bnw_mask)
wc1.recolor(color_func = image_colors)

#For color
wc2.generate(text)
image_colors = ImageColorGenerator(color_mask)
wc2.recolor(color_func = image_colors)
```

### _Step 5 - Plotting the wordcloud_

```python
#Plotting
plt.imshow(wc1, interpolation = 'bilinear')
plt.axis('off')
plt.show()

#Plotting
plt.imshow(wc2, interpolation = 'bilinear')
plt.axis('off')
plt.show()
```

### _Step 6 - Saving the generated wordcloud_

```python
wc1.to_file('batman_bnw_wordcloud.png')
wc2.to_file('batman_color_wordcloud.png')
```
